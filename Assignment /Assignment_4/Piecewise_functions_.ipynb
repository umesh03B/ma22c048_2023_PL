```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-5, 5, 1000001)
```

```python
# Compute the function values
function_values = np.where(x < 0, x * np.sin(x**2), np.cos(x))
```

```python
%timeit function_values
```

```python
# Plot the function
plt.plot(x, function_values)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of the Function')
plt.grid(True)
plt.show()
```
