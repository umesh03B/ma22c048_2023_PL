```python
import urllib.request

# Step 1: Download the Data
url = "ftp://aftp.cmdl.noaa.gov/products/trends/co2/co2_mm_mlo.txt"
data_file = "co2_data.txt"
urllib.request.urlretrieve(url, data_file)
```

```python
import numpy as np

# Step 2: Data Munging
data = np.genfromtxt(data_file)
year = data[:, 0]
month = data[:, 1]
co2_average = data[:, 3]
```

```python
# Step 3: Extract Monthly Averages
co2_monthly_averages = np.ma.masked_invalid(co2_average)
```

```python
import matplotlib.pyplot as plt

# Step 4: Plot Monthly Averages as a Function of Time
dates = year + month / 12.0

plt.figure(figsize=(12, 6))
plt.plot(dates, co2_monthly_averages, linestyle='-', color='b')
plt.xlabel("Time")
plt.ylabel("CO2 Monthly Average")
plt.title("Monthly Averages of Measured CO2")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Step 5: Estimate the Rate of Change of CO2
co2_rate_of_change = np.diff(co2_monthly_averages)
dates_rate_of_change = dates[1:]
```

```python
# Step 6: Plot the Estimate of Rate of Change
plt.figure(figsize=(12, 6))
plt.plot(dates_rate_of_change, co2_rate_of_change, color='c')
plt.xlabel("Time")
plt.ylabel("Rate of Change of CO2")
plt.title("Estimate of Rate of Change of CO2")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
