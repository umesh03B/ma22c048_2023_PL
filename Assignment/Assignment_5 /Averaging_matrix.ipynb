```python
import numpy as np
#generate the matrix with ones on the diagonal, k superdiagonals, and k subdiagonals, and then normalize each row to have a sum of one

def create_averaging_operator(n, k):
    A = np.zeros((n, n))

    for i in range(n):
        for j in range(max(0, i - k), min(n, i + k + 1)):
            A[i, j] = 1

    # Normalize each row
    for i in range(n):
        row_sum=sum(A[i,:])
        A[i, :] /= row_sum

    return A

n = 20
k = 2
A = create_averaging_operator(n, k)
print(A)

```

```python
#Apply A to a vector x whose entries are xj = (-1)^j

x = np.array([(-1)**j for j in range(n)])
result = np.dot(A, x)
print(result)
```

```python
#Apply A to a vector whose entries are values of f(x) = x^2 + 2sin(10x), at n = 1000 equally spaced points in the interval [0, 10], and k = 100:

n = 1000
k = 100
A = create_averaging_operator(n, k)
x = np.linspace(0,10,n)
f_x = x**2 + 2 * np.sin(10 * x)
result = np.dot(A, f_x)
print(result)
```
