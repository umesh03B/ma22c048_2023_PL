```python
import numpy as np

# Define the size of the matrix and grid spacing
n = 1000  # Choose the number of grid points
h = 2 * np.pi / (n - 1)  # Grid spacing


a=np.diag(np.ones(n - 1),k=1)
b=np.diag(-np.ones(n ),k=0)

# Create the differentiation matrix D
D = np.delete(a+b,n-1, 0)

# Define the function f(x) = sin(x) for equally spaced x values
x = np.linspace(0, 2*np.pi, n)

# f_x = np.sin( x)
f_x=np.sin(x)

# Compute the derivative using the differentiation matrix D
df = np.dot(D, f_x/h)

#actual derivative of sin(x)
y=np.cos(x)

# Plot the results
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(x[:-1], df,'--', label="Approx. Derivative",color='blue')
plt.plot(x, y,':',label='cos(x)', color='red')
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.title("Approximate Derivative of f(x) =sin(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define the function f(x) = x^2 + 2sin(10x) for equally spaced x values
f_x_new = x**2 + 2 * np.sin(10 * x)

# Compute the derivative using the differentiation matrix D
df_new = np.dot(D, f_x_new/h)

y1=2*x+20*np.cos(10*x)

# Plot the results

plt.figure(figsize=(10, 6))
plt.plot(x[:-1], df_new, label="Approx. Derivative")
plt.plot(x, y1,':',label='2x+20cos(10x)', color='red')
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.title("Approximate Derivative of f(x) = x^2 + 2sin(10x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
from scipy.sparse import diags

# Define the number of grid points (n) and grid spacing (h)
n = 1000  # Adjust the number of points as needed
h = 2 * np.pi / n  # Grid spacing

# Create the main diagonal (n-1) and the off-diagonal (-1) arrays for D
main_diag = np.ones(n - 1)
off_diag = -np.ones(n - 1)

# Create the differentiation matrix D using scipy's diags function
D_sparse = diags([main_diag, off_diag], [1, 0], shape=(n - 1, n))

# Define the function f(x) for equally spaced x values
x = np.linspace(0, 2 * np.pi, n)

# Define the functions to be differentiated
f_x1 = np.sin(x)
f_x2 = x**2 + 2 * np.sin(10 * x)

# Compute the derivatives using the sparse differentiation matrix D_sparse
df1_sparse = D_sparse.dot(f_x1) / h  # Approximate derivative for sin(x)
df2_sparse = D_sparse.dot(f_x2) / h  # Approximate derivative for x^2 + 2*sin(10x)

# Plot the results
plt.figure(figsize=(16, 6))

plt.subplot(1, 2, 1)
plt.plot(x[:-1], df1_sparse, label="Sparse Approx. Derivative (sin(x))", color='blue')
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.title("Sparse Approximate Derivative of f(x) = sin(x)")
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x[:-1], df2_sparse, label="Sparse Approx. Derivative (x^2 + 2*sin(10x))", color='green')
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.title("Sparse Approximate Derivative of f(x) = x^2 + 2*sin(10x)")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
