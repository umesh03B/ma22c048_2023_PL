```python
import math

# Function to calculate Euclidean distance between two points
def euclidean_distance(point1, point2):
    return math.sqrt((point1[0] - point2[0]) ** 2 + (point1[1] - point2[1]) ** 2)
```

```python
# Function to find k nearest neighbors for each point
def k_nearest(points, k):
    k_nearest_neighbors = []
    for point in points:
      distances=[]
      for other_point in points:
        distances.append((other_point, euclidean_distance(point, other_point)))

      distances.sort(key=lambda x: x[1])  # Sort by distance
      k_nearest = [neighbor[0] for neighbor in distances[1:k + 1]]  # Exclude the point itself
      k_nearest_neighbors.append(k_nearest)
    return k_nearest_neighbors
```

```python
# Given set of points and k
P = [[0, 0], [0.2, 0.22], [0.1, -0.1], [1, 1], [1.1, 0.9], [0.8, 0.9], [1.1, 0.63],
     [0.58, -0.1], [0.63, 0.1], [0.67, -0.3], [0.8, -0.23], [0.8, 0.6]]
k = 3

# Call the function to find k nearest neighbors
k_nearest_neighbors = k_nearest(P, k)
```

```python
# Function to plot arrows with direction
def plot_arrows(points, neighbors):
    for i, point in enumerate(points):
        for neighbor in neighbors[i]:
            plt.annotate("", xy=neighbor, xytext=point,arrowprops={})
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))

# Plot the points
x, y = zip(*P)
plt.scatter(x, y, c='red', marker='o', label='Points')

# Plot arrows with direction from each point to its k-nearest neighbors
plot_arrows(P, k_nearest_neighbors)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('K-Nearest Neighbors')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
