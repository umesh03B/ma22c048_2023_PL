```python
import matplotlib.pyplot as plt   #imports the matplotlib library and renames it as plt.
import numpy as np    #imports the numpy library and renames it as np.
%matplotlib inline

# Create an array of x values in the interval [0, 10]
x = np.linspace(0, 10, 1000)  # You can adjust the number of points for smoother or coarser curve

# Calculate the corresponding y values for sin(x)
y = np.sin(x)

# Plot the graph
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Graph of sin(x)')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# Create a grid of x and y values in the interval [-5, 5] for both x and y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate the corresponding z values for cos(x^2 + y^2)
Z = np.cos(X**2 + Y**2)


# Create a surface plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Customize the plot
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('cos(x^2 + y^2)')
ax.set_title('Surface Plot of cos(x^2 + y^2)')
ax.set_xticks(np.arange(-5, 6, 1))
ax.set_yticks(np.arange(-5, 6, 1))

# Add a color bar to the plot
fig.colorbar(surf, ax=ax, label='cos(x^2 + y^2)')
plt.show()
print("\n")


plt.figure(figsize=(10,7))
plt.imshow(Z, extent=[-5, 5, -5, 5], origin='lower', cmap='viridis')
plt.colorbar(label='cos(x^2 + y^2)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of cos(x^2 + y^2)')
plt.grid(True)
plt.show()
```
