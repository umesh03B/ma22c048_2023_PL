```python
from scipy.linalg import svd # Import the singular value decomposition (SVD) function from SciPy's linear algebra module.
import numpy as np # Import NumPy for numerical computations, and alias it as 'np'.
np.set_printoptions(precision=3, suppress=True)
#Set the print options for NumPy arrays to display numbers with 3 decimal places and suppress small numbers in scientific notation.
```

```python
#Create a random complex matrix 'a' with dimensions 4x5, where both real and imaginary parts are generated randomly.
a = np.random.rand(4, 5) + 1j * np.random.rand(4, 5)
#Perform SVD on matrix 'a', and store the resulting matrices in 'u' (left singular vectors),
# 's' (diagonal singular values), and 'vh' (conjugate transpose of right singular vectors).
u, s, vh = svd(a)
print(a)
print("svd")
print(svd(a))
```

```python
#Calculate the product of 'u' with its conjugate transpose ('u.T.conjugate()').
# This checks whether 'u' is unitary, as the result should be an identity matrix.

u @ u.T.conjugate()   # u is unitary. Its columns are left singulaectr vors
```

```python
#Calculate the product of 'vh' with its conjugate transpose ('vh.T.conjugate()').
#This checks whether the rows of 'vh' are right singular vectors, as the result should be an identity matrix.

vh @ vh.T.conjugate() # Rows of vh are right singular vectors
```

**The geometry of SVD**

```python
a = np.array([[0.1, 0.5], [0.4, 0.8]]) # creates a NumPy array a containing a 2x2 matrix with the specified values.
# computes the Singular Value Decomposition (SVD) of matrix a and
# stores the results in three variables: u, s, and vh. u contains the left singular vectors,
# s contains the singular values, and vh contains the right singular vectors.
u, s, vh = svd(a)
```

```python
#imports the matplotlib library for creating plots and renames it as plt for convenience.
import matplotlib.pyplot as plt
%matplotlib inline

#function to display plots of matrices
def show(c):
    plt.plot(c[0, :], c[1, :])#This line creates a 2D plot of the data in c. It plots
                              #the first row of c as the x-coordinates and the second row of c as the y-coordinates.
    plt.axis('image');


# plot the unit circle and axis segments:

t = np.linspace(0, 3.5 * np.pi , num=300)#generates an array t of 300 evenly spaced values from 0 to 3.5π.

l = np.linspace(-1, 1, num=10)#generates an array l of 10 evenly spaced values from -1 to 1.

z = np.zeros_like(l)#creates an array z filled with zeros and has the same shape as l.

c = np.array([np.concatenate([l, np.cos(t), z]),#onstructs a matrix c by concatenating arrays l, np.cos(t), and z horizontally as the first row,
              np.concatenate([z, np.sin(t), l])])#and concatenating z, np.sin(t), and l horizontally as the second row. This creates the coordinates of points that form a circle.

show(c) #calls the show function to display the circle defined by the matrix c.
```

**Low rank approximation**

```python
# Import necessary libraries and load an image of a cat
cats = plt.imread('/cat_photo.png')
cats.shape#checks and prints the shape of the 'cats' image
```

```python
#calculates the Frobenius norm of the difference between the red (channel 0) and blue (channel 2) color channels of the image
np.linalg.norm(cats[..., 0] - cats[..., 2], 'fro')
```

```python
c = cats[..., 0]
plt.imshow(c, cmap='gray');
# extracts the red color channel from the 'cats' image and displays it as a grayscale image.
```

```python
u, s, vh = svd(c)#Perform Singular Value Decomposition (SVD) on the red color channel
```

```python
plt.plot(s);#Plot the singular values obtained from SVD
```

```python
# Rank 20 approximation of the cats:
l = 20;   cl = u[:, :l] @ np.diag(s[:l]) @ vh[:l, :]
plt.imshow(cl, cmap='gray');
```

```python
# Rank 100 approximation of the cats:
l = 100;   cl = u[:, :l] @ np.diag(s[:l]) @ vh[:l, :]
plt.imshow(cl, cmap='gray');
```

```python
relative_error = 1.e-1
```

```python
#Calculate the relative error to determine the rank for approximation
s2 = s**2
total = np.sum(s2)
diff = np.sqrt((total - np.add.accumulate(s2)) / total)
l = np.argmax(diff < relative_error) + 1
l
```

```python
# Perform the final low-rank approximation of the red color channel

cl = u[:, :l] @ np.diag(s[:l]) @ vh[:l, :]
```

```python
# Calculate the Frobenius norm of the difference between the original and approximated red channel

np.linalg.norm(c - cl, 'fro') / np.linalg.norm(c, 'fro')
```

```python
# Calculate the number of elements in the low-rank representation
u.shape[0] * l + l + l * vh.shape[0]
```

```python
# Calculate the total number of pixels in the original red color channel
c.shape[0] * c.shape[1]
```
